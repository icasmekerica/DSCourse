```python
#import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
#import data
data = pd.read_csv('1.01.Simple-linear-regression.csv')
```

```python
#read data
data
```

```python
#read data
data.describe()
```

```python
#create regression
#define dependenet & independent variables
y = data['GPA']
x1 = data['SAT']
```

```python
#exploring/plotting the data
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPT', fontsize = 20)
plt.show()
```

```python
#regression itself
x = sm.add_constant(x1)
resaults = sm.OLS(y,x).fit()
resaults.summary()
```

```python
#regression line plot
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1, yhat, lw = 4, c = 'orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPT', fontsize = 20)
plt.show()
```

```python
#regression line plot
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1, yhat, lw = 4, c = 'purple', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPT', fontsize = 20)
plt.show()
```
